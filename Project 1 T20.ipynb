# Imported batting summary

```python
df_batting = pd.read_json('t20_wc_batting_summary.json')
```

```python
df_batting.head()
```

```python
batting_array =[]
for i in range(len(df_batting)):
    batting_array.extend(df_batting["battingSummary"][i])
#len(data_batting)
data_batting = pd.DataFrame(batting_array)
len(data_batting)
```

```python
data_batting.head(10)
```

# To add new column and check out and not out

```python
data_batting["Out/Not Out"] = data_batting["dismissal"].apply(lambda x: "Not Out" if x == "" else "Out")
```

```python
data_batting.head(11)
```

# to drop a column from a data set 

```python
data_batting.drop(["dismissal"], axis = 1, inplace = True)
```

```python
data_batting.head(200)
```

# To remove the anomalies 

```python
print(data_batting["batsmanName"][9])
a=data_batting["batsmanName"][9]
a_not = a[-1]
print(a_not)
```

```python
data_batting["batsmanName"]=data_batting["batsmanName"].apply( lambda x: x.replace(a_not,"") )
data_batting.tail(30)
```

```python
print(data_batting["batsmanName"][9])
```

# To creat a new primery key and at it to the new batting table 

```python
match_dict ={}

for index,row in data.iterrows():
    key1 = row["team1"] + " Vs " + row["team2"]
    key2 = row["team2"] + " Vs " + row["team1"]
    match_dict[key1] = row["match_id"]
    match_dict[key2] = row["match_id"]

match_dict
```

```python
data_batting["match_id"] = data_batting["match"].map(match_dict)
data_batting.head()
```

```python
data_batting.to_csv("Batting_Summary.csv", index=False)
```
